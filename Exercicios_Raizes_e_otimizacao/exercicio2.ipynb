```python
from math import sin, cos, tan, atan, e, pi, radians, degrees

def get_phi(_l, _r, _f):
    _phi = atan(2*pi*_f*_l/_r)
    return degrees(_phi)

def func(_beta, _phi):
    _beta_r = radians(_beta)
    _phi_r = radians(_phi)
    res = sin(_beta_r - _phi_r) + sin(_phi_r)*pow(e, -(_beta_r)/tan(_phi_r))
    return res

def dfunc(_beta, _phi):
    _beta_r = radians(_beta)
    _phi_r = radians(_phi)
    res = cos(_beta_r - _phi_r) - pow(e, -_beta_r/tan(_phi_r))*sin(_phi_r)/tan(_phi_r)
    return res

def newton_raphson(phi, x0, tolerance=1e-6, max_iter=3000):
    x = x0
    beta_list = []
    for i in range(max_iter):
        r_func = func(x, phi)
        r_dfunc = dfunc(x, phi)
        x_new = x - r_func / r_dfunc
        beta_list.append(x_new)
        if abs((x_new - x)/x) < tolerance:
            break 
        x = x_new

    return beta_list

# Constantes
f = 1000  # Frequência em Hz
l = 100e-3  # Indutância em Henry
r = 1000  # Resistência em Ohms

phi_val = get_phi(l, r, f)

x0 = 240  # Chute inicial do valor de beta

beta = newton_raphson(phi_val, x0)

print("Ângulo beta para o qual a corrente no diodo se anula:", beta, "graus")
```
